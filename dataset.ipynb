```python
# READ GWDAMA, EXPLORE
# full docs at https://gwnoisehunt.gitlab.io/gwdama/index.html
```

```python
from gwdama.io import GwDataManager

path = '/data2/data2-work-users/mvacatello/data/ML-hackathon/signal_only_dataset.h5'

dama = GwDataManager(path, mode="r") # reading the gwdama object
print(dama)
```

```python
# ACCESS METADATA
```

```python
dama["bbh_only_0_V1_0-2GPS"].show_attrs # prints all metadata associated with the given dataset

print(dama["bbh_only_0_V1_0-2GPS"].attrs["channel"], "\n") # prints a specific metadata, must know metadata label ("channel in this case")

for att in dama["bbh_only_0_V1_0-2GPS"].attrs: # attributes are iterable (same result as .show_attrs method)
    print(att, ":", dama["bbh_only_0_V1_0-2GPS"].attrs[att])
```

```python
# ACCESS DATA, PLOT
```

```python
print(dama["bbh_only_0_V1_0-2GPS"].data) # the actual data, an astropy quantity

dama["bbh_only_0_V1_0-2GPS"].plot() # some fields of this plot are filled with the attributes shown above
```

```python
# LOAD DATA
```

```python
from itertools import islice
import numpy as np

n_samples = 10
dataset = np.empty(shape=(n_samples, 2048)) # rate is 2048 Hz

for idx, signal in enumerate(islice(dama, 0, n_samples), start=0): # dama is iterable, example on the first 10 items
    print(signal) # a string with the name of the dataset
    print(dama[signal].data) # the associated data
    print("\n")
    dataset[idx] = dama[signal].data

print("shape of dataset:", dataset.shape) # embryo of the data to be fed to network
print(dataset)
```
