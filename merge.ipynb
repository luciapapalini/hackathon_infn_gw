```python
import gwdama
from gwdama.io import GwDataManager
import os
import numpy as np
import pandas as pd
import sys, json


# sorting wrt first number in name
import re

def extract_first_number(string):
    # Use regular expression to extract the first number from the string
    match = re.search(r'\d+', string)
    if match:
        return int(match.group())
    else:
        return float('inf')  # Return a large value as default

def sort_array_with_first_number(array):
    # Sort the array based on the first number
    return sorted(array, key=lambda x: extract_first_number(x))

#array = ['bbh_only_125_V1_250-252GPS.hdf5', 'bbh_only_436_V1_872-874GPS.hdf5', 'bbh_only_715_V1_1430-1432GPS.hdf5']
#sorted_array = sort_array_with_first_number(array)
#print(sorted_array)
```

```python
# un dama a partire da un folder

folder_path = '/data2/data2-work-users/mvacatello/data/ML-hackathon/sim_sig_only'  # Replace with the actual folder path

merged_dama = GwDataManager("signal_only_dataset")
attrs_to_dama = False # se gli attributi sono associati al dama
attrs_to_dataset = True # se gli attribut sono associati al dataset

lst = os.listdir(folder_path)
#lst = sort_array_with_first_number(lst)

# Iterate over the files using a for loop
for file_name in lst:
    file_path = os.path.join(folder_path, file_name)
    
    if os.path.isfile(file_path):
        # Perform desired operations on each file
        
        print(file_name)
        with GwDataManager(file_path, mode="r") as temp_dm:
            
            for label in temp_dm:
                merged_dama[temp_dm.attrs["dama_name"]] = temp_dm[label].data # create dataset instance in dama
                
                if attrs_to_dataset == True: # passes attributes from original dataset to target dataset
                    for i in temp_dm[label].attrs:
                        merged_dama[temp_dm.attrs["dama_name"]].attrs[i] = temp_dm[label].attrs[i]
                
                if attrs_to_dama == True: # passes attributes from original dama to target dataset
                    for i in temp_dm.attrs:
                        merged_dama[temp_dm.attrs["dama_name"]].attrs[i] = temp_dm.attrs[i]
            
    print('---\n\n')

print(merged_dama)
```

```python
out_folder = '/data2/data2-work-users/mvacatello/data/ML-hackathon'
merged_dama.write_gwdama(os.path.join(out_folder, "signal_only_dataset.h5"))
```
